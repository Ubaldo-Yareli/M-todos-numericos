Se tiene la función:
$f(x_1 , x_2) = 100 (x_" - x_1^2)^2 + (1-x_1)^2$  
con el punto inicial (iteración 0)
Iteración $x^{(0)} = [-1.2, 1.0] $

```python
import numpy as np

# Punto inicial
x = np.array([-1.2, 1.0], dtype=float)

# Número de iteraciones
n_iter = 2

print("Iteración 0")
print("x =", x)

for k in range(1, n_iter + 1):

    x1, x2 = x
```

Calculo del gradiente
$\nabla f (x_1, x_2)$

```python
    # Gradiente
    g = np.array([
        -400*x1*(x2 - x1**2) + 2*(x1 - 1),
        200*(x2 - x1**2)
    ])
```

Matriz Hessiana $H(x_1, x_2)$


```python
    # Hessiano
    H = np.array([
        [1200*x1**2 - 400*x2 + 2, -400*x1],
        [-400*x1, 200]
    ])
```

Se resuelve el método de Newton con
$Hp = -\nabla f$

```python
    # Paso de Newton: H p = -g
    p = np.linalg.solve(H, -g)
```

$x^{(k+1)} = x ^{(k)} + p^{(k)}$

```python
    # Actualización
    x = x + p
```

k son las iteraciones

```python
    print(f"\nIteración {k}")
    print("Gradiente:")
    print(g)
```

Muestra la matriz Hessiana

```python
    print("\nMatriz Hessiana:")
    print(H)
```

Muestra el vector $p{(k)}$

```python
    print("\nPaso de Newton:")
    print(p)
```

Muestra $x^{(k+1)}$ después de la actualización

```python
    print("\nNuevo punto:")
    print(x)
```

Muestra el valor final de la función

```python
# Valor final de la función
x1, x2 = x
f = 100*(x2 - x1**2)**2 + (1 - x1)**2

print("\nValor final de f =", f)
```

Seguir el método de Newton hasta que converga

```python
tol = 1e-6
max_iter = 100

for k in range(max_iter):

    x1, x2 = x

    g = np.array([
        -400*x1*(x2 - x1**2) + 2*(x1 - 1),
        200*(x2 - x1**2)
    ])

    H = np.array([
        [1200*x1**2 - 400*x2 + 2, -400*x1],
        [-400*x1, 200]
    ])

    p = np.linalg.solve(H, -g)
    x = x + p

    if np.linalg.norm(g) < tol:
        break

print("Solución aproximada:", x)
```

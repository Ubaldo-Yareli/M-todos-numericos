# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x
def g(x):
    return -3**(-x)
def h(x):
    return x-3**(-x)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
#Graficar ambas funciones.
plt.figure(figsize=(7, 7))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.grid()
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
#Mostrarlo.
plt.title('Gr치fica')
plt.xlabel('eje y')
plt.ylabel('eje x')
plt.show()
```

# Gr치fica polinomio de tylor de tercer grado $
f(x)=\sqrt{x+1}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return (1+(1/2)*x)
def h(x):
    return (1+(1/2)*x)-((1/8)*(x**2))
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-1, 10,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x),label="$\sqrt{x+1}$")
plt.plot(x, g(x),label="$\sqrt{(1+(1/2)*x)}$")
plt.plot(x, h(x), label="$(1+(1/2)*x)-((1/8)*(x**2))$")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-1, 10)
plt.ylim(-1, 5)
plt.grid()
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
plt.legend()
plt.title('Gr치fica $f(x)=\sqrt{x+1}$')
plt.xlabel('eje y')
plt.ylabel('eje x')
plt.show()
```

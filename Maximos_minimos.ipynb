```python
import sympy as sp
```

```python
x = sp.symbols('x')
f = x**4 / ((x - 1)*(x - 3)**3)
```

```python
f1 = sp.diff(f, x)
```

```python
print(f"Función f(x): {f} ")
```

```python
print("La primera derivada es :")
print(sp.simplify(f1))
```

```python
puntos_criticos = sp.solve(sp.simplify(f1), x)

print("\nPuntos críticos encontrados:")
print(puntos_criticos)

print("\nAplicando el teorema de suficiencia:")
print("Si f'(x*) = f''(x*) = ... = f^(n-1)(x*) = 0 pero f^(n)(x*) ≠ 0:")
print("- Si n es par y f^(n)(x*) > 0 es un mínimo local")
print("- Si n es par y f^(n)(x*) < 0 es un máximo local")
print("- Si n es impar, no es máximo ni mínimo\n")
```

```python
def clasificar_punto(f, x, punto):
    print(f"\nAnalizando el punto crítico x = {punto}")
    print("Se sabe que f'(x*) = 0.")
    print("Evaluando las derivadas sucesivas hasta encontrar la primera que no sea cero.\n")

    orden = 2

    while True:
        derivada = sp.diff(f, x, orden)
        print(f"\nDerivada de orden {orden}:\n")
        sp.pprint(sp.simplify(derivada))

        valor = sp.simplify(derivada.subs(x, punto))
        print(f"\nLa derivada evaluda en x = {punto} es {valor}")

        if valor == 0:

            print("Según el teorema de suficiencia, se debe seguir derivando ya que aún no se llega a una conclusión.")
            orden += 1
        else:
            print("Se encontró la primera derivada distinta de 0.")

            if orden % 2 == 0:
                print("El orden es PAR.")
                if valor > 0:
                    print("Como es positiva, el punto es un MÍNIMO local.\n")
                    return "mínimo local"
                else:
                    print("Como es negativa, el punto es un MÁXIMO local.\n")
                    return "máximo local"
            else:
                print("El orden es IMPAR.")
                print("Por el teorema de suficiencia, no es máximo ni mínimo.\n")

print("\nClasificación de puntos críticos:")
for punto in puntos_criticos:
    resultado = clasificar_punto(f, x, punto)
    print(f"Conclusión: Hay un {resultado} en x = {punto}")
```
